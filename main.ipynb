```python
import fitz
from tqdm.auto import tqdm

def formatter(text:str) -> str:
    cleaned_text = text.replace('\n',' ').strip()
    return cleaned_text

def openread(path:str) -> list[dict]:
    doc = fitz.open(path)
    pageandtext = []
    for page_number, page in tqdm(enumerate(doc)):
        text = page.get_text()
        text = formatter(text=text)
        text = text.replace("Copyright © 1998 The Bhaktivedanta Book Trust Int\'l. All Rights Reserved. ", "")
        text = text.replace("Kåñëa", "Krishna")
        text = text.replace("kåñëa", "krishna")

        pageandtext.append({"page_number": page_number,
                            "page_char_count": len(text),
                            "page_word_count" : len(text.split(" ")),
                            "page sent_count": len(text.split(". ")),
                            "page_token_count": len(text)/4,
                            "text": text})
    return pageandtext    
pat = openread(path=path)
pat[:2]
```

```python
from spacy.lang.en import English

nlp = English()

nlp.add_pipe("sentencizer")
```

```python
for i in tqdm(pat):
    i["sentences"] = list(nlp(i["text"]).sents)

    i["sentences"] = [str(sentence) for sentence in i["sentences"]]

    i["page_sent_count_sp"] = len(i["sentences"])
    
```

Chunking

```python
sent_chunk_size = 12

def splitter(inp:list[str], slice_size:int=sent_chunk_size) -> list[list[str]]:
    return [inp[i:i +slice_size] for i in range(0, len(inp), slice_size)]

splitter(list(range(50)))
```

```python
for item in tqdm(pat):
    item["sentence_chunks"] = splitter(inp=item["sentences"],
                              slice_size=sent_chunk_size)
    item["no_of_chunks"] = len(item["sentence_chunks"])
```

Independent Chunks

```python
import re

pac = []

for item in tqdm(pat):
    for sent_chunk in item["sentence_chunks"]:
        chunk_dict = {}
        chunk_dict["page_number"] = item["page_number"]

        joined_sent_chunk = "".join(sent_chunk).replace("  ", " ").strip()
        joined_sent_chunk = re.sub(r'\.([A-z])',r'.\1', joined_sent_chunk)

        chunk_dict["sentence_chunk"] = joined_sent_chunk

        chunk_dict["chunk_word_count"] = len([word for word in joined_sent_chunk.split(" ")])
        chunk_dict["chunk_token_count"] = len(joined_sent_chunk)/4

        pac.append(chunk_dict)


len(pac)

        
```
